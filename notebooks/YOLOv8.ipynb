```python
# Import YOLO from ultralytics and load a pre-trained model
from ultralytics import YOLO

# Load a pre-trained YOLOv8s model
model = YOLO('yolov8s.pt')

# Train YOLOv8 model using the dataset
results = model.train(
    data='/content/drive/MyDrive/42174 Artificial Intelligence Studio - Autumn 2025/Dataset/data.yaml', # Path to your dataset config file
    epochs=15,        # Number of training epochs
    imgsz=640,        # Image size used during training (match dataset)
    batch=16          # Batch size (adjust based on GPU memory)
)
```
